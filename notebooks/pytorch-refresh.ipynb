```python
shape = (2,3,)
tensor = torch.rand(shape)
```

```python
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else "cpu"
tensor = tensor.to(device)
tensor.device
```

```python
print(f"Random tensor: \n {tensor}")

tensor[:,2] = 1
print(f"Random tensor now: \n {tensor}")

t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(f"dim1-catted tensor: \n {t1}")

t1 = torch.cat([tensor, tensor, tensor], dim=0)
print(f"dim0-catted tensor: \n {t1}")
```

```python
print(f"Multiplication is element-wise, a priori,\n{tensor * tensor}")
print(f"Mail your Matrix multiplication today!:n{tensor @ tensor.T}")
```

```python
print(f"There is some unsexy tension between in-place operations and derivatives; thus, torch hates this trick. \n {tensor.add_(5)}")
```

```python
n = np.ones(4)
t1 = torch.tensor(n)
t2 = torch.from_numpy(n)
n[0] = 0
np.add(n, 1, out=n)
print(f"Numpy and Tensor love (explicit) sharing!\n {t1} \n {t2}")

t2[0] = 10
print(f"Both ways: \n {t2} \n {n}")
```
