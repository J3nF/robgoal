```julia
import Plots
```

```julia
h(p) = -p * log2(p) - (1 - p) * log2(1 - p)
p = 0.01:0.01:0.99
h1 = h.(p)
Plots.plot(p, h1, label="Entropy", xlabel="p", ylabel="h(p)", title="Entropy Function")

p2 = p.*(1 .- p)
h2 = h.(p2)
Plots.plot!(p, h.(p2), label="Inhibited", xlabel="p*(1-p)", ylabel="h(p)", title="Entropy Function vs p*(1-p)")

Δh = h1 .- h2
Plots.plot!(p, Δh, label="Difference", xlabel="p", ylabel="h(p) - h(p*(1-p))", title="Difference between h(p) and h(p*(1-p))")
```
